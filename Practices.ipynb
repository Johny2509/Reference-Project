```python
#- Cho biết list A gồm bao nhiêu phần tử;
Num_list=[0,5,6,8,22,11,36,39,55,]
print("Number of Charac: ",len(Num_list))
#- Cho biết tổng của các phần tử có trong A;
print("Sum of Charac: ",sum(Num_list))
#- Tìm giá trị của phần tử lớn nhất;
print("Max of Charac: ",max(Num_list))
#- Tìm giá trị của phần tử bé nhất;
print("Min of Charac: ",min(Num_list))
#- Tính trung bình cộng của các phần tử;
print("Average of Charac: {:.2f}".format(sum(Num_list)/len(Num_list)))
#- Sắp xếp list theo thứ tự tăng dần;
print("Ascend of Charac: ",sorted(Num_list))
#- Sắp xếp list theo thứ tự giảm dần;
print("Descend of Charac: ",sorted(Num_list,reverse=True))
```

```python
#W2-L11 Tìm thú trong vườn thú:
#- Tạo ra một list tên có các con thú: animal_name;
#- Nhập vào tên 1 con thú cần tìm: search_name
#- Trả kết quả có tìm thấy được hay không? Nếu tìm thấy thì tìm thấy mấy lần?
animal_name=["bear", "zebra","bear", "pig", "hippo", "elephant","buffallo", "raccon", "crocodile", "monkey", "butterfly", "dragonfly"]
search_name=input("Nhập tên con thú cần tìm trong thảo cầm viên: ")
print("Tên con thú: ",search_name)
for name in animal_name:
    if name==search_name.lower():
        so_lan=animal_name.count(name)
        print(name,"xuất hiện",so_lan,"lần")
        break
else:
    print("Tên con thú không có trong thảo cầm viên!")
```

```python
#W2-L12 Viết chương trình cho phép người dùng nhập vào list số lst_int gồm các phần tử có giá trị tùy ý (nhưng không vượt quá 0-99).
#Máy tính sẽ phát sinh một con số may mắn bất kì: lucky_number trong phạm vi từ  0- 99;
#Nếu lucky_number có trong lst_int thì in ra: "Chúc mừng, bạn đã trúng thưởng!"
#Nếu lucky_number không có trong lst_int thì in ra: "Chúc bạn may mắn lần sau"
import random
so_luot=3
list_int=[]
while so_luot>0:
    X=int(input("Nhập số may mắn của bạn muốn: "))
    if 0<X<100:
        list_int.append(X)
    so_luot-=1
    print("Bạn còn {} lượt thử!".format(so_luot))
lucky_num=random.randint(0,100)
print("Các số may mắn bạn đã chọn: ",list_int)
print("Số may mắn của máy tính:",lucky_num)
if lucky_num in list_int:
    print("Chúc mừng bạn đã trúng thưởng!!!!!!")
else:
    print("Chúc bạn may mắn lần sau T_T")
```

```python
#W2-L13: Cho list gốc: or_lst=[1,5,8,-1,2,6,0,-8,-6,11,19,21,3]
#Tạo new_list bằng cách tăng mỗi phần tử của or_list lên 2 đơn vị;
or_lst=[1,5,8,-1,2,6,0,-8,-6,11,19,21,3]
print("List gốc: ",or_lst)
list_tăng_2=[X+2 for X in or_lst]
print("List sau khi tăng mỗi phần tử hai đơn vị: ",list_tăng_2)
#Tạo new_list bằng cách giảm mỗi phần tử của or_list xuống 2 đơn vị;
list_giảm_2=[X-2 for X in or_lst]
print("List sau khi giảm mỗi phần tử hai đơn vị: ",list_giảm_2)
#Tạo new_list bằng cách lọc ra các phần tử âm trong or_list;
list_âm=[X for X in or_lst if X<0]
print("List sau khi lọc ra các phần từ âm: ",list_âm)
##Tạo new_list bằng cách lọc ra các phần tử dương trong or_list;
list_dương=[X for X in or_lst if X>0]
print("List sau khi lọc ra các phần từ âm: ",list_dương)
#Tạo new_list bằng cách thay các phần tử âm bằng 0;
list_thay_âm=[X if X>=0 else 0 for X in or_lst]
print("List sau khi thay các phần từ âm: ",list_thay_âm)


```

```python
#W2-L16: Viết chương trình quản lý danh bạ điện thoại như sau:
danh_ba={ '0989741258':'Johnny','0903852147':'Katherine','0903712712':'Johnny'}
while True:
    print("Quản Lý Danh Dạ:")
    print("1. Xem Danh Bạ: ")
    print("2. Cập Nhật Lại Tên Mới Theo SĐT: ")
    print("3. Thêm SĐT mới: ")
    print("4. Xóa Liên Hệ Ra Khỏi Danh Bạ: ")
    print("5. Tìm Liên Hệ Trong Danh Bạ: ")

    tac_vu=input("Nhập một tác vụ tương ứng trong bốn tác vụ sau: ")
    if tac_vu=="1":
        print("{:15}{:15}".format("SĐT","Tên"))
        for K,V in danh_ba.items():
            print("{:15}{:15}".format(K,V))
    elif tac_vu=="2":
        SĐT=input("Nhập SĐT cần cập nhật")
        if SĐT in danh_ba:
            ten=input("Nhập tên cần cập nhật: ")
            danh_ba[SĐT]=ten
            print("Cập Nhật Tên Mới Thành Công")
        else:
            print("SĐT chưa tồn tại, vui lòng chọn chức năng thêm mới")
    elif tac_vu=="3":
        SĐT=input("Nhập SĐT cần thêm mới: ")
        if SĐT not in danh_ba:
            ten=input("Nhập tên cần lưu: ")
            danh_ba[SĐT]=ten
            print("Thêm SĐT Mới Thành Công")
        else:
            print("SĐT đã tồn tại, vui lòng chọn chức năng cập nhật")
    elif tac_vu=="4":
        SĐT=input("Nhập SĐT cần xóa: ")
        if SĐT in danh_ba:
            del(danh_ba[SĐT])
            print("Xóa Thành Công SĐT")
        else:
            print("SĐT chưa tồn tại")
    elif tac_vu=="5":
        ket_qua={}
        ten=input("Nhập tên cần tìm: ")
        for K,V in danh_ba.items():
            if ten==V:
                ket_qua[K]=V
        print("Có {} kết quả được tìm thấy.".format(len(ket_qua)))
        if len(ket_qua)>0:
            print("{:15}{:15}".format("SĐT","Tên"))
            for K,V in ket_qua.items():
                 print("{:15}{:15}".format(K,V))
    else:
        print("Bạn đã thoát khỏi chương trình")
        break
```
